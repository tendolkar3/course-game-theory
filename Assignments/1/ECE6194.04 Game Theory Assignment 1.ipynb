Argue why a small cost to the city (by improving public transportation in malnourished areas) can lead to muchhigher utility of its citizens.

### Solution

Malnourished areas generally correspond to areas with low income populations. Their primary means of transportation is public transport. Due to low economic development of the area, facilities like supermarkets, speciality hospitals, employment opportunities are located far away and people living in these regions rely primarily on public transportation for all aspects of their life. Therefore the payoff to every individual, that is added by introducing an extra bus route or increasing bus frequency, is huge. Due to increased convinience, people have access to groceries and medicines, can travel for better jobs and can access speciality medical care.

Every city government has objectives that align with these payoffs of its. For example, the strategic priorities of the Columbus Public Health as stated on their website around the time article was published are -

* Reduce infant mortality
* Reduce overweight and obesity
* Reduce the spread of infectious diseases
* Improve access to public health care
* Implement departmental reorganization

Thus improving public transportation infrastructure in low income areas is a win-win situation for both - the government and its citizens. In game theoretic sense, in a game between citizens in low income areas and the government, the action "use public transportation" is a dominant strategy of citizens, since the citizens in low income areas always depend on it. ## Problem 2.  Strategically Equivalent Games 

Let $ c, c1, c2 \in R $ be scalar constants, and $ 1_{m\times n} $ denote an $ m\times n $ matrix with all entries as 1.  Further, $ 1_m \equiv 1_{m\times 1} $

### Part (a)
Q: Consider a two agent zero-sum game with payoff matrix of Player 1 (the row player) given by $ A\in R_{m\times n} $.  If $ (p^∗, q^∗) $ is a saddle-point equilibrium of the zero-sum game $ (m, n, A) $, then show that $ (p^∗, q^∗) $ is also a saddle-point of the zero-sum game $ (m, n, A+c1_{m\times n}) $.  Hint:  $ 1_{m\times n} = 1_m1^T_n $.

### Solution:
Assume that A is the cost matrix (to be minimized) for player 1 (row player). Therefore, if $ (p^∗, q^∗) $ is the saddle-point equillibrium, we know that - 

$$ p^TAq^* \leq p^{*^T}Aq^* $$ for all $ p $

Now if we change cost matrix $ A $ to $ A+c1_{m\times n} $, the cost incurred for player 1 (for any strategy $ p $) becomes,

$$ p^T(A+c1_{m\times n})q^* = {\color{blue} p^\color{blue}T\color{blue}A\color{blue}q^\color{blue}* \color{blue}+ \color{blue}p^\color{blue}T\color{blue}c\color{blue}1_{\color{blue}m\color{blue}\times \color{blue}n}\color{blue}q^\color{blue}*}$$

Now,

$$ {\color{blue} p^\color{blue}T\color{blue}A\color{blue}q^\color{blue}* \color{blue}+ \color{blue}p^\color{blue}T\color{blue}c\color{blue}1_{\color{blue}m\color{blue}\times \color{blue}n}\color{blue}q^\color{blue}*} \leq {\color{red} p^{\color{red}*^\color{red}T}\color{red}A\color{red}q^\color{red}* \color{red}+ \color{green}p^\color{green}T\color{green}c\color{green}1_{\color{green}m\color{green}\times \color{green}n}\color{green}q^\color{green}*} $$ 

for all $ p $.
For any strategy, pure or mixed, since sum of probabilities is 1, we have,

$$
p^Tc1_{m\times n}q = c
$$

for all $ p $ and $ q $ and therefore,

$$
\color{green}p^\color{green}T\color{green}c\color{green}1_{\color{green}m\color{green}\times \color{green}n}\color{green}q^\color{green}* = \color{red}p^{\color{red}*^\color{red}T}\color{red}c\color{red}1_{\color{red}m\color{red}\times \color{red}n}\color{red}q^\color{red}*
$$

for all p and therefore,

$$ {\color{red} p^{\color{red}*^\color{red}T}\color{red}A\color{red}q^\color{red}* \color{red}+ \color{red}p^{\color{red}*^\color{red}T}\color{red}c\color{red}1_{\color{red}m\color{red}\times \color{red}n}\color{red}q^\color{red}*} = p^{*^T}(A+c1_{m\times n})q^* $$

Therefore,

$$ p^T(A+c1_{m\times n})q^* \leq p^{*^T}(A+c1_{m\times n})q^* $$

for all $ p $. Which means that $ (p^∗, q^∗) $ is the saddle-point equilibrium of the zero-sum game $ (m, n, A+c1_{m\times n}) $. Hence proved.

### Part (b)

for all p and therefore,

Therefore,

$$ p^T(A+c_11_{m\times n})q^* \leq p^{*^T}(A+c_11_{m\times n})q^* $$

#### Part 2 -
Assume that B is the cost matrix (to be minimized) for player 2 (column player). Therefore, if $ (p^∗, q^∗) $ is the Nash equillibrium, we know that - 

$$ p^{*^T}Bq \leq p^{*^T}Bq^* $$ for all $ q $

Now if we change cost matrix $ B $ to $ B+c_21_{m\times n} $, the cost incurred for player 2 (for any strategy $ q $) becomes,

$$ p^{*^T}(B+c_21_{m\times n})q = {\color{blue} p^{\color{blue}*^\color{blue}T}\color{blue}B\color{blue}q \color{blue}+ \color{blue}p^{\color{blue}*^\color{blue}T}\color{blue}c_\color{blue}2\color{blue}1_{\color{blue}m\color{blue}\times \color{blue}n}\color{blue}q}$$

$$ {\color{blue} p^{\color{blue}*^\color{blue}T}\color{blue}B\color{blue}q \color{blue}+ \color{blue}p^{\color{blue}*^\color{blue}T}\color{blue}c_\color{blue}2\color{blue}1_{\color{blue}m\color{blue}\times \color{blue}n}\color{blue}q} \leq {\color{red} p^{\color{red}*^\color{red}T}\color{red}A\color{red}q^\color{red}* \color{red}+ \color{green}p^{\color{green}*^\color{green}T}\color{green}c_\color{green}2\color{green}1_{\color{green}m\color{green}\times \color{green}n}\color{green}q} $$ 

for all $ q $.
For any strategy, pure or mixed, since sum of probabilities is 1, we have,

$$
p^Tc_21_{m\times n}q = c_2
$$

for all $ p $ and $ q $ and therefore,

$$
\color{green}p^{\color{green}*^\color{green}T}\color{green}c_\color{green}2\color{green}1_{\color{green}m\color{green}\times \color{green}n}\color{green}q = \color{red}p^{\color{red}*^\color{red}T}\color{red}c_\color{red}2\color{red}1_{\color{red}m\color{red}\times \color{red}n}\color{red}q^\color{red}*
$$

for all q and substituting,

$$ {\color{red} p^{\color{red}*^\color{red}T}\color{red}B\color{red}q^\color{red}* \color{red}+ \color{red}p^{\color{red}*^\color{red}T}\color{red}c_\color{red}2\color{red}1_{\color{red}m\color{red}\times \color{red}n}\color{red}q^\color{red}*} = p^{*^T}(B+c_21_{m\times n})q^* $$

Therefore,

$$ p^{*^T}(B+c_21_{m\times n})q \leq p^{*^T}(B+c_21_{m\times n})q^* $$

For part1 and part 2, since,

$$ p^{*^T}(B+c_21_{m\times n})q \leq p^{*^T}(B+c_21_{m\times n})q^* $$
for all $ q $ and $$ p^T(A+c_11_{m\times n})q^* \leq p^{*^T}(A+c_11_{m\times n})q^* $$ for all $ p $, we have

$ (p^∗, q^∗) $ is a Nash equilibrium of the non-zero-sum game $ (m, n, A+c_11_{m\times n}, B+c_21_{m\times n}) $. Hence proved.

Prove that if all entries of the matrix $ A+B $ are the same, then the non-zero-sum game is strategically equivalent to a zero-sum game. What would be the payoff matrix of the row player in this strategically equivalent zero-sum game?

### Solution:
Let,
$$ A + B = c1_{m\times n} \text{where, } 1_{m\times n} = 1_m1^T_n $$

From part B, we know that if $ (p^∗, q^∗) $ is a Nash equilibrium of the non-zero-sum game $ (m, n, A, B) $, then $ (p^∗, q^∗) $ is also a Nash equilibrium of the non-zero-sum game $ (m, n, A+c_11_{m\times n}, B+c_21_{m\times n}) $

### Part (a)
Q: Identify a dominated strategy. Formulate another zero-sum game by removing the dominated strategy.

### Solution
Row 3 is the dominated strategy for player 1 (row player) since the payoff by playing row 3 is always lower than payoff by either 1 or 2 irrespective of what other player plays

We can reformulate the zero-sum game by removing the 3rd row since a rational utility maximizing row player will never play row 3 (Or else we have a probability distribution with the third entry as 0)

### Part (b)
Q: Compute the saddle-point equilibrium of the new game without using any calculator/computer aid.

### Solution
The saddle point equillibrium of this zero-sum game with pure strategy does not exist. But there is a mixed strategy equillibrium given by Nash Equillibrium.

To get the NE, we solve two Linear programming problems. Let $p=[a, b]^T$ and $q=[x,y]$. A shortcut way, without the need to solve LP problem would be derived from the fact that, given $p^*$, the opponent (player 2) should be indifferent to playing any of the strategies. Therefore, all greater than zero elements of $p^{*^T}A$ should be equal. Therefore,

$$2a=3b$$

and we also know that $a+b=1$ since they are probabilities. Therefore,
$$ p^* = \begin{bmatrix} \frac{3}{5} \\ \frac{2}{5} \end{bmatrix} $$

Similarly, given $q^*$, the opponent (player 1) should be indifferent to playing any of the strategies. Therefore, all elements of $Aq^{*}$ should be equal. Therefore,

$$2x=3y$$

$$ q^* = \begin{bmatrix} \frac{3}{5} \\ \frac{2}{5} \end{bmatrix} $$

### Part (c)
Q: Let ```m = 100``` and ```n = 400```. Write a code that generates a payoff matrix of the row player using ```A = rand(m,n)```, and computes a saddle-point equilibrium of the two players of the zero-sum game ```(m, n, A)```. You can use the inbuilt ```LPsolverlinprog``` in ```MATLAB``` to solve the problem.

(I will use python instead)

### Solution
The optimisation problem for obtaining the optimal strategy in zero sum games is formulated as (Note that here A is cost matrix, but question asks to consider A as payoff matrix. So negate A in the formulation. But in part C when we generate random numbers in 0 to 1 and negate A, we have to add 1 to make all elements in A positive)

$$ \max_{\widetilde{p}\in R}\sum \widetilde{p_i} $$
Subject to - 
$$ \widetilde{p_i} \geq 0 $$
and
$$ A^T \widetilde{p} \leq 1_n $$

We will use the ```scipy.optimize.linprog``` package for solving this LP problem. In this package, the problem is formulated as -

$$ \min_{x}c^Tx $$
Subject to - 
$$ A_{ub}x \leq b_{ub} 0 $$
and
$$ A_{eq}x = b_{eq} $$

Therefore, for out zero sum game,

<table style="width: 200px; align: center;">
<tr style="background: transparent !important;"><td>$ c = -[1]_m $</td><td>$ A_{ub} = -A^T $</td></tr>
<tr style="background: transparent !important;"><td>$ b_{ub} = [1]_n^T $</td><td>$ x = \tilde{p} $</td></tr>
</table>

Note that a LP problem (as formulated by ```scipy.optimize.linprog```) considers only positive values of variables to be optimized. Therefere the other constraints of $\tilde p \geq 0$ are not necessary.

````python
from scipy.optimize import linprog
import numpy as np
m = 100
n = 400
np.random.seed(10)
c = -np.ones(m)
A = np.random.rand(m,n)
A_ub = np.transpose(A)
b_ub = np.ones((n, 1))
np.seterr(divide='ignore', invalid='ignore')

res_p1 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

c = np.ones(n)
A_ub = -A
b_ub = -np.ones((m, 1))
res_p2 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

print("Number of Iterations for P1 to converge - ",res_p1.nit)
print("p_tilda - \n",res_p1.x)
one_by_v1 = np.sum(res_p1.x)
p_star = res_p1.x/one_by_v1
print("p_star - \n",p_star)

print("Number of Iterations for P2 to converge - ",res_p2.nit)
print("q_tilda - \n",res_p2.x)
one_by_v2 = np.sum(res_p2.x)
q_star = res_p2.x/one_by_v2
print("q_star - \n",q_star)
````

This is the optimal strategy for the row player. If we compute $p^{*^T}A$, all values corresponding to positive p_star elements should be equal (upto a particular tolerance level). This shows that when player 1 plays strategy $p^*$, the opponent is indifferent to playing any strategy. Thus it is a Nash equillibrium. We can verify $p^{*^T}A$ as, 

````python
print(np.dot(p_star,A))
````

We see that for all positive values of $q^*$, the payoffs are not equal! This is because the LP solver did not reach the actual solution. For some reason this solver fails for m = 100 and n = 400. This can also be intuitively seen from the fact that all the lower half of $ q^* $ is 0 which should not be the case for a random distribution within A.

If we use the same solver for a slightly smaller problem, say m = 100 and n = 200, it works fine. Example - 

````python
m = 100
n = 200
np.random.seed(10)
c = -np.ones(m)
A = np.random.rand(m,n)
A_ub = np.transpose(A)
b_ub = np.ones((n, 1))
np.seterr(divide='ignore', invalid='ignore')

res_p1 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

c = np.ones(n)
A_ub = -A
b_ub = -np.ones((m, 1))
res_p2 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

print("Number of Iterations for P1 to converge - ",res_p1.nit)
print("p_tilda - \n",res_p1.x)
one_by_v1 = np.sum(res_p1.x)
p_star = res_p1.x/one_by_v1
print("p_star - \n",p_star)

print("Number of Iterations for P2 to converge - ",res_p2.nit)
print("q_tilda - \n",res_p2.x)
one_by_v2 = np.sum(res_p2.x)
q_star = res_p2.x/one_by_v2
print("q_star - \n",q_star)
````

We see that in this case $ q^* $ looks fairly random. lets evaluate payoff for q when Player 1 plays $ P^* $

## Problem 4.  Fictitious Play and Saddle Point Equilibrium

Let ```m = 10``` and ```n = 30```, and payoff matrix of row player as ```A = rand(m,n)```.   Find its saddle-point equilibrium $(p^∗, q^∗)$ using the code you wrote above.

### Solution
Here, we just change m and n and run the solver again.

````python
#importing again if you want to run just question 4
from scipy.optimize import linprog
import numpy as np
m = 10
n = 30
np.random.seed(1)
c = -np.ones(m)
A = np.random.rand(m,n) # No need to negate and add one since rand and 1-rand are equivalent distributions
A_ub = np.transpose(A)
b_ub = np.ones((n, 1))

res_p1 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

print("Number of Iterations - ",res_p1.nit)
print("p_tilda - \n",res_p1.x)
one_by_v1 = np.sum(res_p1.x)
p_star = res_p1.x/one_by_v1
print("p_star - \n",p_star)
````

Similarly for player 2

````python
c = np.ones(n)
A_ub = np.multiply(-1,A)
b_ub = -np.ones((m, 1))
res_p2 = linprog(c, A_ub, b_ub, 
              options={"disp": False, "tol":1e-12, "bland":True, 
                       "maxiter":10000})

print("Number of Iterations - ",res_p2.nit)
print("q_tilda - \n",res_p2.x)
one_by_v2 = np.sum(res_p2.x)
q_star = res_p2.x/one_by_v2
print("q_star - \n",q_star)
````

````python
actions_of_p1 = [i for i in range(m)]
actions_of_p2 = [i for i in range(n)]

average_payoff_to_p1 = 0

random_generator_for_p1 = np.random.RandomState()
random_generator_for_p2 = np.random.RandomState()
random_generator_for_p1.seed(63)
random_generator_for_p2.seed(36)

history_of_average_payoff_to_p1 = []

for i in range(1,10001):
    current_choice_of_action_for_p1 = random_generator_for_p1.choice(actions_of_p1, 1, p=p_star)
    current_choice_of_action_for_p2 = random_generator_for_p2.choice(actions_of_p2, 1, p=q_star)
#     print(current_choice_of_action_for_p1[0],current_choice_of_action_for_p2[0],A,A[current_choice_of_action_for_p1[0]][current_choice_of_action_for_p2[0]])
    payoff_for_p1 = A[current_choice_of_action_for_p1[0]][current_choice_of_action_for_p2[0]]
    average_payoff_to_p1 = (average_payoff_to_p1*(i-1)+payoff_for_p1)/i
    history_of_average_payoff_to_p1.append(average_payoff_to_p1)
    if i<=3:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==11:
        print("...")
    if i>=98 and i<=100:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==101:
        print("...")
    if i>=498 and i<=500:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==501:
        print("...")
    if i>=998 and i<=1000:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==1001:
        print("...")
    if i>=4998 and i<=5000:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==5001:
        print("...")
    if i>=9998 and i<10000:
        print("GAME ",i)
        print("Player 1 chooses action -",current_choice_of_action_for_p1[0])
        print("Player 2 chooses action -",current_choice_of_action_for_p2[0])
        print("Average payoff for Player 1 -",average_payoff_to_p1)
    if i==10000:
        print("End of silulation")
    
print("Average payoff for Player 1 after 10,000 games -",average_payoff_to_p1)
NE_payoff = np.transpose(p_star).dot(A).dot(q_star)
print("p*T A q* -",NE_payoff)

%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(range(1,len(history_of_average_payoff_to_p1)+1),
         np.subtract(history_of_average_payoff_to_p1,NE_payoff), 
         'b--')
plt.title("Average payoff to Player 1 converging to NE payoff")
plt.xlabel("(Average payoff - NE Payoff) for Player 1")
plt.ylabel("Game Number")
plt.show()
````

We see that the average payoff converges to Nash Equillibrium payoff (both are almost equal)

Let $(\tilde{p}^t,\tilde{q}^t)$ be the empirical distribution tuple of the players at time t. Plot $\Vert\tilde{p}^t−p^∗\Vert$ and $\Vert\tilde{q}^t−q^∗\Vert$ for t ranging from 1 to 10,000. This  procedure  is  called  fictitious  play. Does the empirical distribution seem to converge to the saddle-point equilibrium for this random zero-sum game? 

### Solution
This is interesting. Now instead of knowing the NE probability distributions, The players are calculating empirical probability distribution of opposite player and then reacting to that optimally.

Note that there are 3 ways to choose optimal action of other player given its empirical distribution. Read comment in code.

````python
average_payoff_to_p1 = 0

action_histogram_of_p1 = {actions_of_p1[i]:0 for i in range(m)}
action_histogram_of_p2 = {actions_of_p2[i]:0 for i in range(n)}

history_of_distance_from_optimal_probability_distribution_for_p1 = []
history_of_distance_from_optimal_probability_distribution_for_p2 = []

random_generator_for_p1 = np.random.RandomState()
random_generator_for_p2 = np.random.RandomState()
random_generator_for_p1.seed(0)
random_generator_for_p2.seed(8)

for i in range(1,10001):
    if i==1:
        current_choice_of_action_for_p1 = actions_of_p1[random_generator_for_p1.randint(0,m)]
        current_choice_of_action_for_p2 = actions_of_p2[random_generator_for_p2.randint(0,n)]
        
        action_histogram_of_p1[current_choice_of_action_for_p1]+=1
        action_histogram_of_p2[current_choice_of_action_for_p1]+=1
    else:
        empirical_probability_dist_of_p1 = [
            ((action_histogram_of_p1[actions_of_p1[j]])/(i-1)) 
            for j in range(m)
        ]

        empirical_probability_dist_of_p2 = [
            ((action_histogram_of_p2[actions_of_p2[j]])/(i-1)) 
            for j in range(n)
        ]
        
        '''
        There are 3 ways we could estimate action of P2 given empirical distribution. 
        All 3 ways have different convergence properties. Way 3 converge to Nash Equillibrium. 
        Others converge to different equillibrium points. These points are NOT Nash equillibrium.
        
        Way 1:
            Given empirical_probability_dist_of_p2, make a random choice of pure strategy
        Way 2:
            Choose action of P2 as that index corresponding to maximum in empirical_probability_dist_of_p2
        Way 3:
            Choose action of P2 as mixed strategy i.e. use empirical_probability_dist_of_p2
            
        I have chosen to go with 3rd method. The other 2 methods are commented lines below
        '''
        
        
#         choice_of_p2_as_estimated_by_p1 = random_generator_for_p2.choice(actions_of_p2, 1, p=empirical_probability_dist_of_p2)
#         choice_of_p1_as_estimated_by_p2 = random_generator_for_p1.choice(actions_of_p1, 1, p=empirical_probability_dist_of_p1)
        
#         choice_of_p2_as_estimated_by_p1 = actions_of_p2[np.argmax(empirical_probability_dist_of_p2)]
#         choice_of_p1_as_estimated_by_p2 = actions_of_p1[np.argmax(empirical_probability_dist_of_p1)]
        
#         current_choice_of_action_for_p1 = np.argmin(A[:,choice_of_p2_as_estimated_by_p1])
#         current_choice_of_action_for_p2 = np.argmax(A[choice_of_p1_as_estimated_by_p2,:])
        
        current_choice_of_action_for_p1 = np.argmin(np.dot(A,np.transpose(empirical_probability_dist_of_p2)))
        current_choice_of_action_for_p2 = np.argmax(np.dot(empirical_probability_dist_of_p1,A))
        
        action_histogram_of_p1[actions_of_p1[current_choice_of_action_for_p1]]+=1
        action_histogram_of_p2[actions_of_p2[current_choice_of_action_for_p2]]+=1
        
        payoff_for_p1 = A[current_choice_of_action_for_p1][current_choice_of_action_for_p2]
        
        history_of_distance_from_optimal_probability_distribution_for_p1.append(
            np.sum(np.square(empirical_probability_dist_of_p1-p_star))**(.5)
        )
        
        history_of_distance_from_optimal_probability_distribution_for_p2.append(
            np.sum(np.square(empirical_probability_dist_of_p2-q_star))**(.5)
        )
        
        average_payoff_to_p1 = (average_payoff_to_p1*(i-1)+payoff_for_p1)/i
        if i<=3:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==11:
            print("...")
        if i>=98 and i<=100:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==101:
            print("...")
        if i>=498 and i<=500:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==501:
            print("...")
        if i>=998 and i<=1000:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==1001:
            print("...")
        if i>=4998 and i<=5000:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==5001:
            print("...")
        if i>=9998 and i<10000:
            print("GAME ",i)
            print("Player 1 chooses action -",current_choice_of_action_for_p1)
            print("Player 2 chooses action -",current_choice_of_action_for_p2)
            print("Average payoff for Player 1 -",average_payoff_to_p1)
        if i==10000:
            print("End of silulation")
    
print("Average payoff for Player 1 after 10,000 games -",average_payoff_to_p1)
print("p*T A q* -",np.transpose(p_star).dot(A).dot(q_star))

# plt.plot(range(1,len(history_of_distance_from_optimal_probability_distribution_for_p1)+1),
#          history_of_distance_from_optimal_probability_distribution_for_p1, 
#          'r--', 
#          range(1,len(history_of_distance_from_optimal_probability_distribution_for_p2)+1), 
#          history_of_distance_from_optimal_probability_distribution_for_p2, 
#          'b--')
p1_plot = plt.plot(range(1,len(history_of_distance_from_optimal_probability_distribution_for_p1)+1),
           history_of_distance_from_optimal_probability_distribution_for_p1, 
           'r--', label="Player 1")
p2_plot = plt.plot(range(1,len(history_of_distance_from_optimal_probability_distribution_for_p2)+1),
           history_of_distance_from_optimal_probability_distribution_for_p2, 
           'b--', label="Player 2")
plt.legend(handles=[p1_plot[0], p2_plot[0]])
plt.title("Convergence of fictious play to Nash Equillibrium")
plt.xlabel("Game number")
plt.ylabel("Euclidean distance between \n current mixed strategy and NE strategy")
plt.show()
````

We see that the probability distributions converge to the optimum strategies given by NE. The following two graphs show histogram of probabilities (one given by NE and other one is the converged). This histogram corresponds to the mixed strategy when normalized.
